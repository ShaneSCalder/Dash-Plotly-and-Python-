```python
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
import plotly.figure_factory as ff
from plotly import tools
import numpy as np
```

```python
# Multiple Heat Maps With Subplots
df1 = pd.read_csv('2010SitkaAK.csv')
df2 = pd.read_csv('2010SantaBarbaraCA.csv')
df3 = pd.read_csv('2010YumaAZ.csv')
```

```python
print(df1.columns)
```

```python
Trace1 = go.Heatmap(x=df1['DAY'],
                    y=df1['LST_TIME'],
                    z=df1['T_HR_AVG'],
                    colorscale='Jet',
                    zmin=5,zmax=40)

Trace2 = go.Heatmap(x=df2['DAY'],
                    y=df2['LST_TIME'],
                    z=df2['T_HR_AVG'],
                    colorscale='Jet',
                    zmin=5,zmax=40)

Trace3 = go.Heatmap(x=df3['DAY'],
                    y=df3['LST_TIME'],
                    z=df3['T_HR_AVG'],
                    colorscale='Jet',
                    zmin=5,zmax=40)

# if you set shared_yaxes to false each subplot will have time shown to the left of each plot
fig = tools.make_subplots(rows=1,cols=3,
                          subplot_titles=['Sitka AK', 'SB CA', 'Yuma AZ'],
                          shared_yaxes=True)
# to Set a title for subplots 
# fig['layout'].update(title='Temps for 3 cities')

fig.append_trace(Trace1,1,1)
fig.append_trace(Trace2,1,2)
fig.append_trace(Trace3,1,3)

pyo.plot(fig)

```
