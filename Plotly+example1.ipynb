```python
import numpy as np
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
np.random.seed(42)
```

```python
random_x = np.random.randint(1,101,100)
```

```python
random_y = np.random.randint(1,101,100)
```

```python
data = [go.Scatter
        (x = random_x,
         y = random_y,
         mode = 'markers',
         marker=dict(
             size=12,
             color='rgb(51,204,153)',
             symbol='pentagon',
             line = {'width':2}))]
```

```python
layout = go.Layout(title='Hello My First PLot', xaxis={'title':'My X Axis'}, yaxis=dict(title='MY Y Axis'), hovermode='closest')
fig = go.Figure(data=data,layout=layout)
pyo.plot(fig, filename = 'scatter.html')
```
